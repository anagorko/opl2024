    > The story of Gomory cuts is characterized by swings between great acclaim in
    the early days, near oblivion for decades and an amazing comeback in the last
    20 years. These cuts have been described as “elegant”, “disappointing” and
    “the clear winner” at various times over the last 55 years. This essay retraces
    that roller coaster. Ralph Gomory’s paper “Early Integer Programming” recounts 
    his discovery of fractional cuts. It is a few years after he wrote his doctoral dissertation on
    nonlinear differential equations that he heard of linear programming for the
    first time.  He was working for the Navy at the time.  In one particular instance, 
    it would have been preferable to have solutions in integers. Gomory
    thought that, somehow, one should be able to accomplish this. Within a few
    days he had invented fractional cuts. His approach was to first solve the linear
    program and then, using appropriate integer linear forms, to generate valid
    linear inequalities cutting off the undesirable fractional
    solution. By adding these cuts to the linear program, solving again using the simplex algorithm
    and iterating, Gomory could solve by hand any small integer linear program
    that he tried. However, he did not have a finiteness proof yet. At this point, he happened to run into Martin Beale in the halls of Princeton
    University in late 1957 and mentioned that he could solve linear programs in integers. When
    Beale immediately responded “but that’s impossible”, Gomory realized that he was not the first to think about this problem. 
    As it turns out, Dantzig, Fulkerson, and Johnson had pioneered the cutting plane approach in a seminal paper published in 1954. 
    They devised special-purpose cuts for the traveling salesman problem and, as a result, were able to solve to optimality an instance
    with 48 cities. However, Gomory’s goal was different and more ambitious. His fractional cuts were general-purpose cuts that applied to all integer linear programs.

Jeżeli $x^\ast$ jest rozwiązaniem optymalnym relaksacji problemu całkowitoliczbowego, naruszającym któryś z warunków całkowitoliczbowości, to dowolną nierówność, która
  nie jest spełniona przez $x^\ast$ a jest spełniona

## Cięcia Gomory'ego

Dany jest problem optymalizacyjny

$c^T x \to \max \\
Ax \leq b, \\
x \geq 0 \\
x \in \mathbb{Z}^n$

liniowy z dodatkowymi **nieliniowymi** warunkami $x \in \mathbb{Z}^n$. Uwaga: zakładamy, że **wszystkie** zmienne mają przyjąć wartości całkowite i wszystkie współczynniki problemu są całkowite ($A \in M_{m\times n}(\mathbb{Z})$, $b \in \mathbb{Z}^m$, $c \in \mathbb{Z}^n$). Za pomocą metody sympleks rozwiązujemy **relaksację** problemu

$
c^T x \to \max \\
Ax \leq b, \\
x \geq 0
$

znajdując zbiór bazowy optymalny $\mathcal{B}$ i rozwiązanie bazowe optymalne $x^\ast \in \mathbb{R}^n$ takie, że

$
x^\ast_\mathcal{B} = A^{-1}_\mathcal{B} b\\
x^\ast_\mathcal{N} = 0.
$

Znamy też wartość funkcji celu:

$
\xi = \xi^\ast - (c^\ast)^T x_\mathcal{N}.
$

(Dla pewnych wartości $\xi^\ast$, $c^\ast$ których tu nie specyfikujemy.)



Jeżeli $x^\ast \in \mathbb{Z}^n$, to rozwiązaliśmy dany problem. W przeciwnym przypadku wybieramy indeks $i \in \mathcal{B}$ taki, że $x^\ast_i \not\in \mathbb{Z}$.
W tabelce mamy wiersz

$
x_i = b^\ast_i - \sum_{j \in \mathcal{N}} a^\ast_{ij} x_j.
$

Oczywiście $b^\ast_i$ nie jest całkowite. Jeżeli wszystkie współczynniki $a^\ast_{ij}$ są nieujemne, to oczywiście każde rozwiązanie problemu spełnia nierówność

$
x_i \leq b^\ast_i
$

a każde całkowitoliczbowe rozwiązanie spełnia nierówność

$
x_i \leq \lfloor b^\ast_i \rfloor.
$

Jest to cięcie - ta nie równość nie odcina żadnych rozwiązań całkowitoliczbowych problemu, a odcina rozwiązanie $x^\ast$. Oczywiście zazwyczaj niektóre współczynniki $a^\ast_{ij}$ będą ujemne. Wtedy możemy przepisać nasze równanie:

$
x_i = b^\ast_i - \sum_{j \in \mathcal{N}} a^\ast_{ij} x_j = b^\ast_i - \sum_{j \in \mathcal{N}} (\lfloor a^\ast_{ij} \rfloor + \{ a^\ast_{ij} \}) x_j
$

gdzie $\{ x \}$ to taka liczba, że $x = \lfloor x \rfloor + \{ x \}$. Oczywiście $\{ x \} \geq 0$ dla każdego $x \in \mathbb{R}$. Wtedy

$ 
x_i \leq \lfloor b^\ast_i \rfloor - \sum_{j \in \mathbb{N}} \lfloor a^\ast_{ij} \rfloor x_j
$

Tę nierówność nazywamy **cięciem Gomory'ego**. Uwaga: jeżeli pojawiają się tu zmienne dodane przy przejściu do postaci równościowej, to zastępujemy je zmiennymi oryginalnego problemu używając podstawień z oryginalnej tabelki.

Cięcie jest poprawne, bo

$
x_i + \sum_{j \in \mathcal{N}} \lfloor a^\ast_{ij} \rfloor x_j = b^\ast_i - \sum_{j \in \mathbb{N}} \lfloor a^\ast_{ij} \rfloor x_j \leq b^\ast_i
$

i lewa strona nierówności musi przyjmować wartość całkowitą dla $x \in \mathbb{Z}^n$ (tu pokazuje się konieczność założenia, że wszystkie zmienne $x_i$ mają być całkowite, w tym zmienne dodane - do tego potrzebna jest całkowitoliczbowość $A$ i $b$).

Skoro $c \in \mathbb{Z}^n$, to analogicznie możemy utworzyć cięcie z równości

$
\xi = \xi^\ast - (c^\ast)^T x_\mathcal{N},
$

bo wiemy, że wartość $\xi$ musi być całkowita.
Tu akurat (skoro $x^\ast$ było rozwiązaniem optymalnym) cięciem będzie po prostu

$
\xi \leq \lfloor \xi^\ast \rfloor.
$

## Przykład

```python
A = matrix(QQ, 2, 2, [3, -1, 3, 11])
b = matrix(QQ, 2, 1, [12, 66])
c = matrix(QQ, 1, 2, [3, 4])

P = InteractiveLPProblemStandardForm(A,b,c)
show(P)
show(plot(P, xmin=0, xmax=7, ymin=0, ymax=7))
```

```python
P.run_simplex_method()
```

```python
# ... generujemy cięcia dla wszystkich trzech równości ...

P1 = P.add_constraint([1, 0], 5)
P1 = P1.add_constraint([3, 4], 34)
show(P1)
show(plot(P1, xmin=0, xmax=7, ymin=0, ymax=7))
```

```python
show(P1.run_simplex_method())
show(P1.initial_dictionary())
```

```python
# x2 = 51/11- ... x4 + x5 - ... x5
# x2 - x5 <= 51/11
# x2 - x5 <= 4
# x2 - 5 + x1 <= 4
# x1 + x2 <= 9

P2 = P1.add_constraint([1, 1], 9)
show(P2)
show(plot(P2, xmin=0, xmax=7, ymin=0, ymax=7))
```

```python
show(P2.run_simplex_method())
show(P2.initial_dictionary())
```

```python
# x1 - x4 <= 4
# x1 - 66 + 3x1 + 11x2 <= 4
# 4x1 + 11x2 <= 70

P3 = P2.add_constraint([4, 11], 70)
show(P3)
show(plot(P3, xmin=0, xmax=7, ymin=0, ymax=7))
```

```python
show(P3.run_simplex_method())
show(P3.initial_dictionary())
```

```python
# x2 - x7 <= 4
# x2 - 9 + x1 + x2 <= 4
# x1 + 2x2 <= 13

P4 = P3.add_constraint([1, 2], 13)
show(P4)
show(plot(P4, xmin=0, xmax=7, ymin=0, ymax=7))
```

```python
show(P4.run_simplex_method())
show(P4.initial_dictionary())
```
