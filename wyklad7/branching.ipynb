# Strong branching

```python
# podział po x1

P1a = P.add_constraint([1, 0], 2)
P1b = P.add_constraint([-1, 0], -3)

print('1a: ', P1a.optimal_value(), P1a.optimal_solution())
print('1b: ', P1b.optimal_value(), P1b.optimal_solution())
```

```python
# podział po x2

P2a = P.add_constraint([0, 1], 1)
P2b = P.add_constraint([0, -1], -2)

print('2a: ', P2a.optimal_value(), P2a.optimal_solution())
print('2b: ', P2b.optimal_value(), P2b.optimal_solution())
```
