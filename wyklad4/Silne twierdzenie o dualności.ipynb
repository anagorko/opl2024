# Silne twierdzenie o dualności

Mamy dane dwa problemy.

<table>
    <tr>
        <td>
            <strong>Problem pierwotny.</strong>
$$
\sum_{j=1}^n c_j x_j \to \max
$$
$$
\sum_{j=1}^n a_{ij} x_j \leq b_i \ \ \ \ 1 \leq i \leq m
$$
$$
x_j \geq 0 \ \ \ \ 1 \leq j \leq n
$$
        </td>
        <td>
<strong>Problem dualny.</strong>
$$
\sum_{i=1}^m b_i y_i \to \min
$$
$$
\sum_{i = 1}^m y_i a_{ij} \geq c_j \ \ \ \ 1 \leq j \leq n
$$
$$
y_i \geq 0 \ \ \ \ 1 \leq i \leq m
$$
        </td>
    </tr>
</table>

[Przypominam](http://slahaie.net/docs/lpdual.pdf), że problem dualny bierze się z przemnożenia nierówności problemu pierwotnego stronami przez zmienne $y_i$, dodania ich stronami i nałożenia warunków, przy których wartość funkcji celu problemu dualnego jest górnym ograniczneniem na wartość funkcji celu problemu pierwotnego: to słabe twierdzenie o dualności. Jednakże powyższe sformułowanie należy traktować jako definicję problemu dualnego do problemu pierwotnego danego w postaci standardowej.

**Silne twierdzenie o dualności**

Jeżeli problem pierwotny ma rozwiązanie optymalne $$x^\ast = (x_1^\ast, x_2^\ast, \ldots, x_n^\ast)$$ to problem dualny ma rozwiązanie optymalne $$y^\ast = (y_1^\ast, y_2\ast, \ldots, y_m^\ast)$$ takie, że $$\sum_{j=1}^n c_j x_j^\ast = \sum_{i=1}^m b_i y_i^\ast,$$ tzn. jeżeli problem pierwotny ma rozwiązanie optymalne, to problem dualny też ma rozwiązanie optymalne i optymalne wartości funkcji celu obu problemów są równe.

## Analiza na przykładzie

Rozważmy problem liniowy w postaci standardowej
$$
\begin{array}{l}
\begin{array}{lcrcrcrcl}
 \max \mspace{-6mu}&\mspace{-6mu}  \mspace{-6mu}&\mspace{-6mu} 4 x_{1} \mspace{-6mu}&\mspace{-6mu} + \mspace{-6mu}&\mspace{-6mu} x_{2} \mspace{-6mu}&\mspace{-6mu} + \mspace{-6mu}&\mspace{-6mu} 3 x_{3} \mspace{-6mu}&\mspace{-6mu}  \mspace{-6mu}&\mspace{-6mu} \\
 \mspace{-6mu}&\mspace{-6mu}  \mspace{-6mu}&\mspace{-6mu} x_{1} \mspace{-6mu}&\mspace{-6mu} + \mspace{-6mu}&\mspace{-6mu} 4 x_{2} \mspace{-6mu}&\mspace{-6mu}  \mspace{-6mu}&\mspace{-6mu}  \mspace{-6mu}&\mspace{-6mu} \leq \mspace{-6mu}&\mspace{-6mu} 1 \\
 \mspace{-6mu}&\mspace{-6mu}  \mspace{-6mu}&\mspace{-6mu} 3 x_{1} \mspace{-6mu}&\mspace{-6mu} - \mspace{-6mu}&\mspace{-6mu} x_{2} \mspace{-6mu}&\mspace{-6mu} + \mspace{-6mu}&\mspace{-6mu} x_{3} \mspace{-6mu}&\mspace{-6mu} \leq \mspace{-6mu}&\mspace{-6mu} 3 \\
\end{array} \\
x_{1}, x_{2}, x_{3} \geq 0
\end{array}
$$
Problem dualny to
$$
\begin{array}{l}
\begin{array}{lcrcrcl}
 \min \mspace{-6mu}&\mspace{-6mu}  \mspace{-6mu}&\mspace{-6mu} y_{1} \mspace{-6mu}&\mspace{-6mu} + \mspace{-6mu}&\mspace{-6mu} 3 y_{2} \mspace{-6mu}&\mspace{-6mu}  \mspace{-6mu}&\mspace{-6mu} \\
 \mspace{-6mu}&\mspace{-6mu}  \mspace{-6mu}&\mspace{-6mu} y_{1} \mspace{-6mu}&\mspace{-6mu} + \mspace{-6mu}&\mspace{-6mu} 3 y_{2} \mspace{-6mu}&\mspace{-6mu} \geq \mspace{-6mu}&\mspace{-6mu} 4 \\
 \mspace{-6mu}&\mspace{-6mu}  \mspace{-6mu}&\mspace{-6mu} 4 y_{1} \mspace{-6mu}&\mspace{-6mu} - \mspace{-6mu}&\mspace{-6mu} y_{2} \mspace{-6mu}&\mspace{-6mu} \geq \mspace{-6mu}&\mspace{-6mu} 1 \\
 \mspace{-6mu}&\mspace{-6mu}  \mspace{-6mu}&\mspace{-6mu}  \mspace{-6mu}&\mspace{-6mu}  \mspace{-6mu}&\mspace{-6mu} y_{2} \mspace{-6mu}&\mspace{-6mu} \geq \mspace{-6mu}&\mspace{-6mu} 3 \\
\end{array} \\
y_{1}, y_{2} \geq 0
\end{array}
$$

Tabelka problemu pierwotnego (po dodaniu zmiennych $w_1, w_2$ i wybraniu bazy $\{ w_1, w_2 \}$) to
$$
\renewcommand{\arraystretch}{1.5} %notruncate
\begin{array}{rcrcrcrcr}
\zeta \mspace{-6mu}&\mspace{-6mu} = \mspace{-6mu}&\mspace{-6mu} 0 \mspace{-6mu}&\mspace{-6mu} + \mspace{-6mu}&\mspace{-6mu} 4 x_{1} \mspace{-6mu}&\mspace{-6mu} + \mspace{-6mu}&\mspace{-6mu} x_{2} \mspace{-6mu}&\mspace{-6mu} + \mspace{-6mu}&\mspace{-6mu} 3 x_{3}\\
\hline
w_{1} \mspace{-6mu}&\mspace{-6mu} = \mspace{-6mu}&\mspace{-6mu} 1 \mspace{-6mu}&\mspace{-6mu} - \mspace{-6mu}&\mspace{-6mu} x_{1} \mspace{-6mu}&\mspace{-6mu} - \mspace{-6mu}&\mspace{-6mu} 4 x_{2} \mspace{-6mu}&\mspace{-6mu}  \mspace{-6mu}&\mspace{-6mu} \\
w_{2} \mspace{-6mu}&\mspace{-6mu} = \mspace{-6mu}&\mspace{-6mu} 3 \mspace{-6mu}&\mspace{-6mu} - \mspace{-6mu}&\mspace{-6mu} 3 x_{1} \mspace{-6mu}&\mspace{-6mu} + \mspace{-6mu}&\mspace{-6mu} x_{2} \mspace{-6mu}&\mspace{-6mu} - \mspace{-6mu}&\mspace{-6mu} x_{3}\\
\end{array}
$$

Tabelka problemu dualnego (po dodaniu zmiennych $z_1, z_2, z_3$ i wybraniu bazy $\{ z_1, z_2, z_3 \}$) to
$$
\renewcommand{\arraystretch}{1.5} %notruncate
\begin{array}{rcrcrcr}
-\xi \mspace{-6mu}&\mspace{-6mu} = \mspace{-6mu}&\mspace{-6mu} 0 \mspace{-6mu}&\mspace{-6mu} - \mspace{-6mu}&\mspace{-6mu} y_{1} \mspace{-6mu}&\mspace{-6mu} - \mspace{-6mu}&\mspace{-6mu} 3 y_{2}\\
\hline
z_{1} \mspace{-6mu}&\mspace{-6mu} = \mspace{-6mu}&\mspace{-6mu} -4 \mspace{-6mu}&\mspace{-6mu} + \mspace{-6mu}&\mspace{-6mu} y_{1} \mspace{-6mu}&\mspace{-6mu} + \mspace{-6mu}&\mspace{-6mu} 3 y_{2}\\
z_{2} \mspace{-6mu}&\mspace{-6mu} = \mspace{-6mu}&\mspace{-6mu} -1 \mspace{-6mu}&\mspace{-6mu} + \mspace{-6mu}&\mspace{-6mu} 4 y_{1} \mspace{-6mu}&\mspace{-6mu} - \mspace{-6mu}&\mspace{-6mu} y_{2}\\
z_{3} \mspace{-6mu}&\mspace{-6mu} = \mspace{-6mu}&\mspace{-6mu} -3 \mspace{-6mu}&\mspace{-6mu}  \mspace{-6mu}&\mspace{-6mu}  \mspace{-6mu}&\mspace{-6mu} + \mspace{-6mu}&\mspace{-6mu} y_{2}\\
\end{array}
$$
Proszę zwrócić uwagę, że o ile zbiór bazowy wybrany dla problemu pierwotnego jest dopuszczalny, to zbiór bazowy problemu dualnego ($z_1, z_2, z_3$) **nie jest dopuszczalny**. Byłby to problem, gdybyśmy chcieli na tej tabelce uruchomić algorytm sympleks: nie będziemy tego robić. Z drugiej strony **jest to prawidłowa** reprezentacja problemu dualnego sprowadzonego do postaci równościowej.

Wypiszmy same współczynniki obu problemów (uwzględniając funkcje celu zapisane na górze) w postaci macierzy:
$$
\left(\begin{array}{rrrr}
0 & 4 & 1 & 3 \\
1 & -1 & -4 & 0 \\
3 & -3 & 1 & -1
\end{array}\right), \quad\quad\quad
\left(\begin{array}{rrr}
0 & -1 & -3 \\
-4 & 1 & 3 \\
-1 & 4 & -1 \\
-3 & 0 & 1
\end{array}\right)
$$
Zwróćmy uwagę, że prawa macierz (problemu dualnego) to minus transpozycja lewej macierzy (problemu pierwotnego).

Uruchamiamy metodę sympleks dla problemu pierwotnego (którego zbiór bazowy jest dopuszczalny). Jako zmienną wchodzącą do bazy wybieramy $x_3$, wychodzącą $w_2$ i otrzymujemy tabelkę
$$

\hline
w_{1} \mspace{-6mu}&\mspace{-6mu} = \mspace{-6mu}&\mspace{-6mu} 1 \mspace{-6mu}&\mspace{-6mu} - \mspace{-6mu}&\mspace{-6mu} x_{1} \mspace{-6mu}&\mspace{-6mu} - \mspace{-6mu}&\mspace{-6mu} 4 x_{2} \mspace{-6mu}&\mspace{-6mu}  \mspace{-6mu}&\mspace{-6mu} \\
x_{3} \mspace{-6mu}&\mspace{-6mu} = \mspace{-6mu}&\mspace{-6mu} 3 \mspace{-6mu}&\mspace{-6mu} - \mspace{-6mu}&\mspace{-6mu} 3 x_{1} \mspace{-6mu}&\mspace{-6mu} + \mspace{-6mu}&\mspace{-6mu} x_{2} \mspace{-6mu}&\mspace{-6mu} - \mspace{-6mu}&\mspace{-6mu} w_{2}\\
\end{array}
$$

\hline
w_{1} \mspace{-6mu}&\mspace{-6mu} = \mspace{-6mu}&\mspace{-6mu} 1 \mspace{-6mu}&\mspace{-6mu} - \mspace{-6mu}&\mspace{-6mu} x_{1} \mspace{-6mu}&\mspace{-6mu} - \mspace{-6mu}&\mspace{-6mu} 4 x_{2} \mspace{-6mu}&\mspace{-6mu}  \mspace{-6mu}&\mspace{-6mu} \\
w_{2} \mspace{-6mu}&\mspace{-6mu} = \mspace{-6mu}&\mspace{-6mu} 3 \mspace{-6mu}&\mspace{-6mu} - \mspace{-6mu}&\mspace{-6mu} 3 x_{1} \mspace{-6mu}&\mspace{-6mu} + \mspace{-6mu}&\mspace{-6mu} x_{2} \mspace{-6mu}&\mspace{-6mu} - \mspace{-6mu}&\mspace{-6mu} x_{3}\\

Na tabelce problemu dualnego wykonamy operację "dualną": w problemie pierwotnym zmienną wchodzącą było $x_3$ (trzecia zmienna bazy); w problemie dualnym zmienną wychodzącą będzie $z_3$ (trzecia zmienna zbioru niebazowego). W problemie pierwotnym zmienną wychodzącą było $w_2$ (druga zmienna zbioru niebazowego); w problemie dualnym zmienną wchodzącą będzie $y_2$ (druga zmienna bazy). Proszę zwrócić uwagę, że w zapisie macierzowym współczynników obu problemów kolumny lewej macierzy odpowiadają kolejno: wyrazowi wolnemu, $x_1$, $x_2$, $x_3$ a wiersze: funkcji celu, $w_1$, $w_2$; kolumny prawej macierzy odpowiadając kolejno: wyrazowi wolnemu, $y_1$, $y_2$; a wiersze: funkcji celu, $z_1$, $z_2$, $z_3$. Dobór zmiennych wchodzących/wychodzących w tabelce problemu dualnego to wybór zmiennych odpowiadających zmiennym problemu pierwotnego przy transpozycji macierzy.

\hline
z_{1} \mspace{-6mu}&\mspace{-6mu} = \mspace{-6mu}&\mspace{-6mu} 5 \mspace{-6mu}&\mspace{-6mu} + \mspace{-6mu}&\mspace{-6mu} y_{1} \mspace{-6mu}&\mspace{-6mu} + \mspace{-6mu}&\mspace{-6mu} 3 z_{3}\\
z_{2} \mspace{-6mu}&\mspace{-6mu} = \mspace{-6mu}&\mspace{-6mu} -4 \mspace{-6mu}&\mspace{-6mu} + \mspace{-6mu}&\mspace{-6mu} 4 y_{1} \mspace{-6mu}&\mspace{-6mu} - \mspace{-6mu}&\mspace{-6mu} z_{3}\\

Wykonajmy kolejny krok metody sympleks dla problemu pierwotnego: wchodzi $x_2$, wychodzi $w_1$, oraz wykonajmy "dualny krok" dla problemu dualnego: wchodzi $y_1$, wychodzi $z_2$. Uzyskujemy tabelki

\end{array}
$$

\hline
z_{1} \mspace{-6mu}&\mspace{-6mu} = \mspace{-6mu}&\mspace{-6mu} 6 \mspace{-6mu}&\mspace{-6mu} + \mspace{-6mu}&\mspace{-6mu} \frac{1}{4} z_{2} \mspace{-6mu}&\mspace{-6mu} + \mspace{-6mu}&\mspace{-6mu} \frac{13}{4} z_{3}\\

\end{array}

Znaleźliśmy rozwiązanie optymalne problemu pierwotnego: wszystkie współczynniki funkcji celu są niedodatnie. Jednocześnie w problemie dualnym znaleźliśmy pierwsze rozwiązanie dopuszczalne: wszystkie współczynniki w kolumnie wyrazów wolnych są nieujemne. Cały czas (w każdej tabelce) wartości funkcji celu na znalezionych rozwiązaniach były równe (tylko rozwiązania problemu dualnego nie były dopuszczalne).
W szczególności znalezione rozwiązanie dopuszczalne problemu dualnego ma wartość funkcji celu równą optymalnej wartości funkcji celu problemu pierwotnego. Ze słabego twierdzenia o dualności wnioskujemy, że jest to wartość optymalna (co widać też po współczynnikach funkcji celu w tabelce problemu dualnego).

## Dowód silnego twierdzenia o dualności

Uogólniając powyższą procedurę do dowodu potrzeba nam tylko dwóch faktów:

1. Wykonując kroki algorytmu sympleks ("pivoty") zawsze znajdziemy rozwiązanie optymalne problemu pierwotnego.

2. Wykonując "kroki dualne" w problemie dualnym zachowamy własność "macierz współczynników tabelki problemu dualnego to minus macierz transpononowana do macierzy współczynników tabelki problemu pierwotnego".

Pierwszy fakt jest łatwy: z założenia problem pierwotny ma rozwiązanie optymalne, więc jeżeli będziemy wykonywali pivoty zgodnie z regułą Blanda, to na pewno znajdziemy rozwiązanie optymalne. (W przykładzie nie stosowałem reguły Blanda by szybciej dojść do celu: ważne dla nas jest tylko to, że jakaś ścieżka do celu istnieje.)

Aby uzasadnić drugi fakt trzeba sprawdzić co dzieje się z macierzą współczynników przy wykonywaniu jednego pivotu:

$$
\begin{array}{|c|c|c|}
\hline
 & & \\
 \hline
 \quad\quad\quad b \quad\quad\quad & a & \\
 \hline
 d \rule[-9mm]{0mm}{20mm} & c \\
 \hline
\end{array}
\qquad \underrightarrow{\text{pivot}} \qquad
\begin{array}{|c|c|c|}
\hline
 & & \\
 \hline
 \quad\quad\quad -\frac ba \rule[-3mm]{0mm}{8mm} \quad\quad\quad & \frac 1a & \\
 \hline
 d - \frac{bc}a \rule[-9mm]{0mm}{20mm} & \frac{c}{a} \\
 \hline
\end{array}
$$

Macierz współczynników problemu dualnego to minus transpozycja macierzy problemu pierwotnego, zaś zmienne wchodzące/wychodzące są transponowane:
$$
\begin{array}{|c|c|c|}
\hline
  & -b & -d \rule[-9mm]{0mm}{20mm}  \\
  \hline
  & -a & \quad\quad\quad -c \quad\quad\quad \\
 \hline
 & & \\
 \hline
\end{array}
\qquad \underrightarrow{\text{pivot}} \qquad
\begin{array}{|c|c|c|}
\hline
  & \frac ba & -d + \frac{bc}a \rule[-9mm]{0mm}{20mm}  \\
  \hline
  & -\frac 1a & \quad\quad\quad -\frac ca \rule[-3mm]{0mm}{8mm} \quad\quad\quad \\
 \hline
 & & \\
 \hline
\end{array}
$$

Czyli rzeczywiście własność "minus transpozycji" jest zachowana, co kończy dowód.
