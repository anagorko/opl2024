```python
from sage.plot.plot3d.plot3d import axes
from sage.plot.plot3d.shapes import Sphere
from sage.plot.plot3d.shapes import LineSegment
```

```python
p = polygon3d([[1,0,0], [0,1,0], [0,0,1]], color=(0.8, 0.8, 0.8), alpha=0.8, frame=False)
p += Sphere(0.015, color='black').translate(1,0,0)
p += Sphere(0.015, color='black').translate(0,1,0)
p += Sphere(0.015, color='black').translate(0,0,1)
p += LineSegment((1,0,0), (0,1,0), color='black', thickness=0.15)
p += LineSegment((1,0,0), (0,0,1), color='black', thickness=0.15)
p += LineSegment((0,1,0), (0,0,1), color='black', thickness=0.15)
```

```python
x = plot(p + axes(1.25, 0.25, color=(0,0,0)))
```

```python
show(x, frame=False, viewpoint=([-0.1263,-0.4504,-0.8839],152.16))
```

```python
k = 1.25
A = vector([0,1,0]) + k * vector([1, 1, 0])
D = vector([0,0,1]) + k * vector([1,0,1])
p = polygon3d([A, [0,1,0], [0,0,1], D], color=(0.8, 0.8, 0.8), alpha=0.8, frame=False)
p += Sphere(0.015, color='black').translate(-1,0,0)
p += Sphere(0.015, color='black').translate(0,1,0)
p += Sphere(0.015, color='black').translate(0,0,1)
p += LineSegment((0,1,0), A, color='black', thickness=0.15)
p += LineSegment((0,0,1), D, color='black', thickness=0.15)
p += LineSegment((0,1,0), (0,0,1), color='black', thickness=0.15)
p += LineSegment((-1.25, 0, 0), (0,0,0), color='black', thickness=0.1)
p += LineSegment((-1, 0, 0), (0,1,0), color='black', thickness=0.05)
p += LineSegment((-1, 0, 0), (0,0,1), color='black', thickness=0.05)
```

```python
show(p + axes(1.25, 0.25, color=(0,0,0)), viewpoint=([-0.9254,-0.2662,-0.2697],93.69))
```

```python
k = 1.25
A = vector([1,0,0]) + k * vector([0, 1, 0])
D = vector([1,0,0]) + k * vector([0, 1, 1])
C = vector([1,0,0]) + k * vector([0, 0, 1])
p = polygon3d([A, [1,0,0], C, D], color=(0.8, 0.8, 0.8), alpha=0.8, frame=False)
p += Sphere(0.015, color='black').translate(1,0,0)
p += LineSegment((1,0,0), A, color='black', thickness=0.15)
p += LineSegment((1,0,0), C, color='black', thickness=0.15)
```

```python
show(p + axes(1.25, 0.25, color=(0,0,0)), viewpoint=([-0.9254,-0.2662,-0.2697],93.69))
```

```python
k = 0.25
p = Sphere(0.015, color='black').translate(1,1,0)
p += Sphere(0.015, color='black').translate(0,1,1)
p += LineSegment((1,1, 0), (0,1,1), color='black', thickness=0.15)
p += LineSegment((1,1, 0), vector([1,1,0]) + k * vector([1,0,-1]), color='black', thickness=0.05)
p += LineSegment((0,1, 1), vector([0,1,1]) - k * vector([1,0,-1]), color='black', thickness=0.05)
```

```python
show(p + axes(1.25, 0.25, color=(0,0,0)), viewpoint=([-0.9835,-0.1153,-0.1395],80.11), frame=False)
```

```python
k = 0.25
A = vector([0, 1, 1])
B = vector([1, 0, 0])
p = Sphere(0.015, color='black').translate(A)
p += Sphere(0.015, color='black').translate(B)
p += LineSegment(A, B, color='black', thickness=0.15)
p += LineSegment(A, A + k * (A - B), color='black', thickness=0.05)
p += LineSegment(B, B + k * (B - A), color='black', thickness=0.05)
```

```python
show(p + axes(1.25, 0.25, color=(0,0,0)), viewpoint=([-0.9835,-0.1153,-0.1395],80.11), frame=False)
```
