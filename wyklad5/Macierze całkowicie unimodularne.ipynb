## Macierze całkowicie unimodularne

**Definicja.** Mówimy, że macierz $A$ jest **całkowicie unimodularna** jeżeli wyznacznik każdej kwadratowej podmacierzy $A$ (powstałej z $A$ przez wybór $k$ wierszy i $k$ kolumn) jest równy $0$, $1$ lub $-1$.

Uwaga: podmacierz może być wymiaru $1 \times 1$, czyli w szczególności jeżeli $A$ jest całkowicie unimodularna, to wszystkie jej elementy należą do zbioru $\{-1, 0, 1\}$.

**Lemat 1.** Niech $A$ będzie macierzą całkowicie unimodularną i niech $\widetilde A$ będzie macierzą powstałą z $A$ przez dodanie wektora jednostkowego $e_i$ jako ostatniej kolumny. Wtedy $\widetilde A$ jest całkowicie unimodularna.

**Dowód.** Ustalmy podmacierz $Q$ rozmiaru $l \times l$ macierzy $\widetilde A$. Jeżeli $Q$ jest podmacierzą $A$, to $\det(Q) \in \{-1, 0, 1\}$ z całkowitej unimodularności $A$. W przeciwnym przypadku liczymy $\det(Q)$ za pomocą rozwinięcia Laplace'a względem ostatniej kolumny macierzy $Q$. Mamy

$$
\det(Q) = \sum_{i=1}^l (-1)^{i+l} q_{il} \det(Q^{il}),
$$
gdzie $Q^{il}$ to macierz powstała z $Q$ poprzez usunięcie $i$-tego wiersza i $l$-tej (ostatniej) kolumny.
Z założenia, $l$-ta kolumna macierzy $Q$ to albo wektor zerowy (wtedy $\det(Q) = 0$) albo jest tam dokładnie jeden niezerowy element $q{kl}=1$. W drugim przypadku mamy

$$
\det(Q) = (-1)^{k+l} \det(Q^{kl}) \in \{ -1, 0, 1 \},
$$
bo $Q^{kl}$ jest podmacierzą $A$.

**Lemat 2.** Rozpatrzmy problem liniowy zawierający $n$ nieujemnych zmiennych $x_1, x_2, \ldots, x_n$ i $m$ nierówności postaci

$$
c^T x \to \max
$$
$$
Ax \leq b
$$
$$
x \geq 0,
$$

gdzie $b \in \mathbb{Z}^m$. Jeżeli $A$ jest całkowicie unimodularna i jeżeli problem liniowy ma rozwiązanie optymalne, to ma też rozwiązanie optymalne $x^\ast \in \mathbb{Z}^n$.

**Dowód.** Zacznijmy od sprowadzenia problemu do postaci równościowej. Po dodaniu zmiennych uzyskujemy układ równań $\bar A \bar x = b$, gdzie $\bar A = (A | I_m)$ ($I_m$ - macierz jednostkowa $m \times m$) i $\bar x \in \mathbb{R}^{n+m}$. Rozwiązujemy problem (w postaci równościowej) za pomocą metody sympleks. Niech $\bar x^\ast$ będzie rozwiązaniem bazowym optymalnym problemu, dla bazy $\mathcal{B} \subset \{ 1, 2, \ldots, n + m \}$. Oczywiście niezerowe elementy wektora $\bar x^\ast$ równe są
$$
\bar x^\ast_\mathcal{B} = \bar A^{-1}_\mathcal{B} b.
$$
Ze wzoru Cramera elementy macierzy $\bar A^{-1}_\mathcal{B}$ można zapisać w postaci ułamków, ze wspólnym mianownikiem $\det(\bar A_\mathcal{B})$. Macierz $\bar A_\mathcal{B}$ jest kwadratową podmacierzą macierzy $\bar A$, która jest całkowicie unimodularna z całkowitej unimodularności macierzy $A$ i Lematu 1. Ponieważ $\bar A_\mathcal{B}$ jest nieosobliwa, mamy $\det(\bar A_\mathcal{B}) \in \{ -1, 1 \}$. Wynika stąd, że $\bar x^\ast \in \mathbb{Z}^{n+m}$, co daje całkowitoliczbowe rozwiązanie optymalne danego problemu liniowego.

**Uwaga.** Udowodniliśmy, że każde rozwiązanie bazowe danego problemu sprowadzonego do postaci równościowej jest całkowitoliczbowe.


## Macierze sąsiedztwa grafów dwudzielnych są całkowicie unimodularne

Niech $G = (X \sqcup Y, E)$ będzie grafem dwudzielnym ($\sqcup$ oznacza sumę rozłączną; dwudzielność oznacza, że każda krawędź grafu łączy wierzchołek z $X$ z wierzchołkiem z $Y$) o $n$ wierzchołkach $v_1, v_2, \ldots v_n$ i $m$ krawędziach $e_1, e_2, \ldots, e_m$.
**Macierz sąsiedztwa** grafu $G$ to macierz $A \in M_{n \times m}(\mathbb{R})$ o elementach zdefiniowanych wzorem
$$
a_{ij} = \left\{
\begin{array}{ll}
1 & \text{ dla } v_i \in e_j \\
0 & \text{ w przeciwnym przypadku}.
\end{array}
\right.
$$

**Przykład.** Macierz sąsiedztwa pełnego grafu dwudzielnego o $3 + 4$ wierzchołkach. Wiersze odpowiadają wierzchołkom, kolumny - krawędziom. Jak widać pierwsza kolumna odpowiada krawędzi łączącej wierzchołek $0$ z wierzchołkiem $3$.

```python
G = graphs.CompleteBipartiteGraph(3, 4)
show(G)
print(G.incidence_matrix())
```

**Lemat 3.** Jeżeli $G = (X \sqcup Y, E)$ jest grafem dwudzielnym, to macierz sąsiedztwa $A$ grafu $G$ jest całkowicie unimodularna.

**Dowód.** Musimy udowodnić, że każda podmacierz $Q$ rozmiaru $l \times l$ macierzy $A$ ma wyznacznik $0$, $1$ lub $-1$. Dowód prowadzimy przez indukcję po $l$.
Przypadek $l=1$ jest prosty, bo macierz sąsiedztwa z definicji zawiera tylko elementy $0$ i $1$.

Weźmy $l > 1$ i ustalmy podmacierz $Q$ rozmiaru $l\times l$ macierzy $A$. Ponieważ kolumny $Q$ odpowiadają krawędziom grafu, każda kolumna zawiera co najwyżej dwa niezerowe elementy (równe $1$). Jeżeli macierz $Q$ zawiera kolumnę samych zer, to $\det(Q) = 0$, co kończy dowód. Jeżeli macierz $Q$ zawiera kolumnę z dokładnie jednym niezerowym elementem, to rozwinięcie Laplace'a względem tej kolumy pokazuje, że badany wyznacznik jest równy $\pm$ wyznacznikowi podmacierzy $(l-1)\times (l-1)$, która z założenia indukcyjnego ma wyznacznik równy $0, 1$ lub $-1$.

Pozostaje do rozpatrzenia przypadek, gdy każda kolumna $Q$ zawiera dokładnie dwie jedynki. Pokażemy, że wtedy $\det(Q) = 0$. Rzeczywiście, suma wszystkich wierszy $Q$ odpowiadających wierzchołkom z $X$ to wektor $(1, 1, 1, \ldots, 1)$. Podobnie suma wszystkich wierszy $Q$ odpowiadających wierzchołkom z $Y$ to wektor $(1, 1, \ldots, 1)$. Skoro zbiory $X$ i $Y$ są rozłączne, pokazuje to liniową zależność wierszy macierzy $Q$, czyli jej wyznacznik jest równy $0$.

## Minimalne pokrycie wierzchołkowe jako problem całkowitoliczbowy

Jak wyżej, graf $G = (X \sqcup Y, E)$ jest grafem dwudzielnym o wierzchołkach $v_1, v_2, \ldots, v_n$ i krawędziach $e_1, e_2, \ldots, e_m$.

Niech $A$ będzie macierzą sąsiedztwa grafu $G$. Rozważmy problem
$$
\sum_{i = 1}^n y_i \to \min
$$
$$
A^T y \geq 1
$$
$$
y \geq 0
$$
$$
y \in \mathbb{Z}^n.
$$

W tym problemie każdy wiersz macierzy $A^T$, odpowiadający krawędzi $e_j$, zadaje warunek
$$
\sum_{i \colon v_i \in e_j} y_i \geq 1.
$$
Wynika stąd, że każdym rozwiązaniu *optymalnym* $\bar y$ mamy $\bar y_i \in \{0,1\}$ dla każdego $i$, bo większe wartości można zmniejszyć do $1$ bez naruszenia warunków problemu (zmniejszając przy tym wartość funkcji celu). Wtedy wierzchołki grafu $v_i$ dla których $\bar y_i = 1$ tworzą minimalne pokrycie wierzchołkowe grafu $G$.

## Maksymalne skojarzenie jako problem całkowitoliczbowy

Jak wyżej, graf $G = (X \sqcup Y, E)$ jest grafem dwudzielnym o wierzchołkach $v_1, v_2, \ldots, v_n$ i krawędziach $e_1, e_2, \ldots, e_m$.

Niech $A$ będzie macierzą sąsiedztwa grafu $G$. Rozważmy problem
$$
\sum_{j=1}^m x_j \to \max
$$
$$
A x \leq 1
$$
$$
x \geq 0
$$
$$
x \in \mathbb{Z}^m.
$$

W tym problemie każdy wiersz macierzy $A$, odpowiadający wierzchołkowi $v_i$, zadaje warunek
$$
\sum_{j \colon e_j \ni v_i} x_j \leq 1.
$$

Wynika stąd, że $x_j \in \{ 0, 1 \}$ i jeżeli $\bar x$ jest optymalnym rozwiązaniem, to wybór krawędzi $e_j$ dla których $\bar x_j = 1$ zadaje maksymalne skojarzenie w grafie $G$.

## Dowód twierdzenia Kőniga

Zauważmy, że **relaksacje** problemów odpowadających problemowi znalezenia maksymalnego skojarzenia i minimalnego pokrycia są problemami liniowymi dualnymi. Z silnego twierdzenia o dualności ich rozwiązania są równe. Ale macierze $A$ i $A^T$ w obu problemach są całkowicie unimodularne z Lematu 3. Wobec tego z Lematu 2, możemy znaleźć optymalne rozwiążania problemów zrelaksowanych, które są całkowitoliczbowe. Stąd wielkość maksymalnego skojarzenia w grafie dwudzielnym jest równa wielkości minimalnego pokrycia wierzchołkowego.
