```python
#
# Dane problemu przepływu sieci zadane są za pomocą grafu
#

G = Graph([("o","a",3), ("o","b",1), ("o","c",1),("a","b",1),("a","d",1), \
           ("b","e",3), ("c","d",4), ("c","e",4),("d","n",4),("e","n",1)])

source = "o"; # nazwa wierzchołka początkowego
target = "n"; # nazwa wierzchołka końcowego

G.plot(edge_labels=True).show()

# Sposób odczytania wag krawędzi
for u in G.neighbors('a'):
    print('a --', u, G.edge_label('a',u))
```
