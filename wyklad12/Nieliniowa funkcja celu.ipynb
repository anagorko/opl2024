# Nieliniowa funkcja celu

Zobaczymy jak za pomocą zmiennych o wartościach całkowitych można modelować problemy o kawałkami liniowej funkcji celu.



Oczywiście dowolną nieliniową funkcję celu możemy aproksymować funkcją liniową, ma to więc zastosowanie do dowolnych nieliniowych funkcji.



Zmienną $x$ rozbijamy na sumę
$$
x = x_1 + x_2 + \cdots + x_k,
$$
gdzie zmienna $x_i$ koduje jak duża część $i$-tego przedziału dziedziny mieści się w przedziale $[0, x]$. 
Oczywiście w tym przedziale mieści się część początkowych segmentów w całości, kolejny częściowo, a następne wcale.

Chcemy wprowadzić warunki, które gwarantują takie wartościowanie zmiennych.
$$
L_j w_j \leq x_j \leq L_j w_{j-1}, \ \ \ j = 1, 2, \ldots ,k
$$
$$
w_0 = 1
$$
$$
w_j \in \{ 0, 1 \}, \ \ \ j = 1, 2, \ldots ,k
$$
$$
x_j \geq 0, \ \ \ j = 1, 2, \ldots ,k
$$

Rzeczywiście, pierwszy warunek wymusza nierówność $w_j \leq w_{j-1}$.

Wszystkie zmienne $x_j$ dla których $w_j = 1$ muszą być równe $L_j$, bo pierwszy warunek przyjmuje postać $L_j \leq x_j \leq L_j$.

Wszystkie zmienne $x_j$, dla których $w_{j-1} = 0$ muszą być równe $0$, bo pierwszy warunek przyjmuje postać $0 \leq x_j \leq 0$.

Jest jedna zmienna, dla której $w_{j - 1} = 1$ i $w_j = 0$ - wtedy zmienna $w_j$ spełnia nierówność $0 \leq x_j \leq L_j$.

O takie warunki nam chodziło. Teraz kawałkami liniową funkcję celu możemy zapisać jako

$$
K + c_1 x_1 + c_2 x_2 + \cdots + c_k x_k
$$
